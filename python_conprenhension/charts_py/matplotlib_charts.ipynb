```python
#import matplotlib

import matplotlib.pyplot as plt

def generate_bar_chart():
    labels =['a','b','c']
    values = [100, 200, 300,]
    fig, ax = plt.subplots()
    ax.bar(labels,values)
    plt.show()

if __name__ == "__main__":
    generate_bar_chart()

    
```

```python
# de manera dinamica  con un pie 

import matplotlib.pyplot as plt

def generate_pie_chart(labels, values):
    fig,ax =plt.subplots()
    ax.pie(values, labels=labels)
    ax.axis("equal")

if __name__ == "__main__":
    labels =['gasto','comida','invest']
    values =[200,240,720]
    generate_pie_chart(labels,values)
```
