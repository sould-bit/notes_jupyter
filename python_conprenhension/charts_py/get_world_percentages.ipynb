```python
import csv
from csv import DictReader

def read(path) -> str:
    with open(path, "r+" )as csvfile:
        read = csv.DictReader(csvfile, delimiter=",")
        data = []
        for datecsv in read:
            # datecsv = tuple(datecsv.items())
            data.append(datecsv)
        return data
        
            


```

```python
date = read("date.csv")
def get_world(date):
    # for select in date:
    #      return select[16]
    # print(select)

    # selec = [porcentage[16] for porcentage in date]
    # for tuple in selec:
    #     worl = tuple[0]
    #     porcen = tuple[1]
    #     return worl,porcen
    continet = [item["Continent"] for item in date]
    quest = input("eligue un continente , :"+ str(continet))

    date = [item for item in date if item["Continent"] == quest]

    selet = {item["Country/Territory"]: item["World Population Percentage"]for item in date}
    worl= selet.keys()
    perc = selet.values()
    return worl, perc
    
```

```python
import matplotlib.pyplot as plt
def generate_chart_pie(labels, values):
    fig, ax = plt.subplots()
    ax.pie(values, labels =labels)
    ax.axis("equal")


if __name__ == "__main__":
    data = read("date.csv")
    labels, values = get_world(data)
    generate_chart_pie(labels, values)
```
